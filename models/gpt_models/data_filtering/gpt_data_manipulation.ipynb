```python
import pandas as pd
import json

# Sample TSV data as a string
data = "data/hate_int_prof_SVO.tsv"

# Convert the TSV string to a pandas DataFrame
df = pd.read_csv((data), sep='\t')

# Extract only the required columns
df = df[['Sentence', 'Intensity']]

# Construct the desired JSON structure
result = []

for index, row in df.iterrows():
    entry = {
        "id": index + 1,
        "sentence": row["Sentence"],
        "intensity": row["Intensity"]
    }
    result.append(entry)

# Convert the result to JSON
json_result = json.dumps(result, indent=4)

# Print the JSON result
print(json_result)
```

```python
with open('data/hate_intensity.json', 'w') as f:
    f.write(json_result)
```

```python
import json
import random



# Load the JSON data
data = json.loads(json_result)

# Organize data by intensity
intensity_dict = {}
for item in data:
    intensity = int(item["intensity"])
    if intensity not in intensity_dict:
        intensity_dict[intensity] = []
    intensity_dict[intensity].append(item)

# Get 20 sentences for each intensity from 1 to 10
result = {}
for i in range(1, 11):
    if i in intensity_dict:
        result[i] = random.sample(intensity_dict[i], min(20, len(intensity_dict[i])))
    else:
        result[i] = []

# Print the result
for intensity, items in result.items():
    for item in items:
        print(items)
```

```python
hate_intensity_prompt_samples = json.dumps(result, indent=4)
```

```python
with open('data/hate_intensity_prompt_samples.json', 'w') as f:
    f.write(hate_intensity_prompt_samples)
```
